<h3>Logistic Regression using gradient descent</h3>

```python
from sklearn.datasets import make_classification
import numpy as np
X,y = make_classification(n_samples=100,n_features=2,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=20,
                          n_informative=1,n_redundant=0)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],cmap="winter",s=100,c=y)
```

<h3>sigmoid function</h3>

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

    
```

<h3>gradient descent</h3>

```python
def gradient_descent(X,y):
    X = np.insert(X,0,1,axis = 1)
    weights = np.ones(X.shape[1])
    lr = 0.5
    for i in range(1000):
        y_hat = sigmoid(np.dot(X,weights))
        weights = weights + lr*(np.dot((y-y_hat),X)/X.shape[0])
    return weights[1:],weights[0]
```

```python
coef,intercept = gradient_descent(X,y)  
```

```python
b = -(intercept/coef[1])
m = -(coef[0]/coef[1])
```

```python
x_input = np.linspace(3,-3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='black',linewidth=2)
plt.scatter(X[:,0],X[:,1],cmap="winter",s=100)
plt.ylim(-3,2)
```
