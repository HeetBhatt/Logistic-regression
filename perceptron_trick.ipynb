<h3>perceptron trick</h3>

```python
from sklearn.datasets import make_classification
import numpy as np
X,y = make_classification(n_samples=100,n_features=2,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=20,
                          n_informative=1,n_redundant=0)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],cmap="winter",s=100,c=y)
```

```python
def step(h):
    if h>0:
        return 1
    return 0
```

```python
def perceptron(X,y):
    lr = 0.1
    X=np.insert(X,0,1,axis = 1)
    weights = np.ones(X.shape[1])
    for i in range(1000):
        x = np.random.randint(0,100)
        y_hat = step(np.dot(X[x],weights))
        weights = weights + lr*(y[x]-y_hat)*X[x]
    return weights[0],weights[1:]
        
```

```python
intercept,coef = perceptron(X,y)
print(intercept,coef)
```

```python
#ax + by + c = 0
#m = -a/b
#c = -c/b
m = -coef[0]/coef[1]
b = -intercept/coef[1]
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='black',linewidth=2)
plt.scatter(X[:,0],X[:,1],cmap="winter",s=100)
plt.ylim(-3,2)
```

<h3>sigmoid function</h3>

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

    
```

```python
def perceptron1(X,y):
    lr = 0.1
    X=np.insert(X,0,1,axis = 1)
    weights = np.ones(X.shape[1])
    for i in range(1000):
        x = np.random.randint(0,100)
        y_hat = sigmoid(np.dot(X[x],weights))
        weights = weights + lr*(y[x]-y_hat)*X[x]
    return weights[0],weights[1:]
        
```

```python
intercept,coef = perceptron1(X,y)
print(intercept,coef)
```

```python
#ax + by + c = 0
#m = -a/b
#c = -c/b
m = -coef[0]/coef[1]
b = -intercept/coef[1]
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='black',linewidth=2)
plt.scatter(X[:,0],X[:,1],cmap="winter",s=100)
plt.ylim(-3,2)
```
